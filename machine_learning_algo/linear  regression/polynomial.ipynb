## polynomial  Regression

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
## this are imp library 
```

```python
x = 6*np.random.rand(100,1) -3
y = 0.5*x**2 +1.5*x+2+np.random.rand(100 ,1)
```

```python
plt.scatter(x,y,color = 'g')
plt.xlabel("x dataset")
plt.ylabel("y datasets")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## for cheaking we go to do simple linear regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_1= LinearRegression()
```

```python
reg_1.fit(x_train , y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test , reg_1.predict(x_test))
print(score)
```

```python
plt.plot(x_train, reg_1.predict(x_train), color = 'r')
plt.scatter(x_train,y_train)
plt.xlabel("x dataset")
plt.ylabel("y dataset ")
```

```python
# lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree= 3 , include_bias= True)
x_train_Poly = poly.fit_transform(x_train)
x_test_Poly = poly.transform(x_test)
```

```python
from sklearn.metrics import r2_score
reg = LinearRegression()
reg.fit(x_train_Poly, y_train)

y_pred = reg.predict(x_test_Poly)

score = r2_score(y_test, y_pred)
print("R2 Score :", score)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x_train , reg.predict(x_train_Poly) , color = "green")
plt.scatter(x_train, y_train)
plt.xlabel("x datasets ")
plt.ylabel("y datasets")
```

```python
# prediction of new data
x_new = np.linspace(-3,3,200).reshape(200,1)
x_new_poly = poly.transform(x_new)
```

```python
y_new = reg.predict(x_new_poly)

plt.plot(x_new, y_new, "r-", linewidth=2, label="Prediction")
plt.plot(x_train, y_train, "b.", label="Training Points")
plt.plot(x_test, y_test, "g.", label="Test Points")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
print("how much our model is accurate :" , score)
```
