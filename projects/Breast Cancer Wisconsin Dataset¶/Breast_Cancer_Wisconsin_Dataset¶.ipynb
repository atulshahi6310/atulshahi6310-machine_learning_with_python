## Loading a dataset from sklearn 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder 
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
```

```python
## find relation btw 0	Malignant 1	Benign with respect to other 
plt.figure(figsize=(12,12))
plt.subplot(2, 2, 1)
sns.boxplot(x='target', y='mean radius', data=df)
plt.ylabel('Radius Mean')
plt.xlabel('Diagnosis')
plt.title('Dianosis vs Radius Mean')
plt.subplot(2,2,2)
sns.boxplot(x ='target' , y= "mean texture" , data = df)
plt.title('Dianosis vs mean texture')
plt.subplot(2,2,3)
sns.boxplot(x = 'target' , y ='mean perimeter', data = df)
plt.subplot(2,2,4)
sns.boxplot(x = 'target' , y= 'mean area' , data =df)
plt.title('dianosis vs mean area')
```

From the above figures and visualizations, we see that those with malignant tumors had higher values for 'mean radius', 'mean texture', 'mean perimeter', 'mean area', and 'mean smoothness' compared to those with benign tumors.
